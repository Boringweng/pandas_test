```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  
```

```python
dead105 = pd.read_csv('dead105.csv')
dead106 = pd.read_csv('dead106.csv')
air = pd.read_csv('aircsv.csv')
county = pd.read_csv('county.csv')
#讀取CSV

temp_county_dict = county.to_dict()
#csv to dict

county_dict = dict()
air_dict = dict()
#建立dict()格式空值

air.dropna(inplace=True)
#取除air多餘的空值

for a in air.values :
    air_dict.update({a[0]:a[8]})
#將各縣市 AQI>100日數的比率放入 air_dict

for c in temp_county_dict['name'].items() :
    county_dict.update({temp_county_dict["county"][c[0]] : c[1][:3]})
#將各縣市 county 代碼與 縣市名放入並取去掉鄉鎮區
```

```python
lung_cancer = dead106[dead106['cause']==21]
#選取肺癌的死亡人口

counties = list(lung_cancer["county"].values)
#取 lung_cancer 中的城市的代碼 並放入list

cities_name = list()
#建立用來存城市名稱的 list

index = list(lung_cancer.index)
#儲存第一筆肺癌的index

for c in counties :
    try :
        cities_name.append(county_dict[c])
    except :
        cities_name.append(np.nan)
lung_cancer["city"] = pd.Series(cities_name, index=index)
#在最後方新增了一欄 城市名稱

cities = list(lung_cancer["city"].values)
#取全部的城市名稱

AQI = list()
#建立用來存AQI值的空 list

for c in cities :
    try :
        AQI.append(air_dict[c])
    except :
        AQI.append(0)      
lung_cancer["AQI"] = pd.Series(AQI, index=index)
#比對並新增一欄個城市的API數值
    
lung_cancer.head()
```

```python
from sklearn.linear_model import LinearRegression #lm 複迴歸

from sklearn import preprocessing, linear_model #logistic_regr        Logistic 迴歸模型
from sklearn.feature_selection import f_regression

from sklearn import cross_validation, ensemble, preprocessing, metrics# bag    Bagging
```

```python
#lung_cancer = lung_cancer.reset_index()
feature_cols = ['sex', 'age_code' ,'AQI'] #將資料塞入模型中
X =lung_cancer[feature_cols]
y = lung_cancer.N

lm = LinearRegression()
lm.fit(X, y)

logistic_regr = linear_model.LogisticRegression()
logistic_regr.fit(X,y)

label_encoder = preprocessing.LabelEncoder()
bag = ensemble.BaggingClassifier(n_estimators = 100)
bag_fit = bag.fit(X,y)
```

```python
lm_predictions = lm.predict(X) #複迴歸準確率
accuracy = lm.score(X,y)
print(accuracy)
```

```python
log_predictions = logistic_regr.predict(X) # Logistic 迴歸準確率
accuracy = logistic_regr.score(X,y)
print(accuracy)
```

```python
test_y_predicted = bag.predict(X)  #Bagging 準確率
accuracy = metrics.accuracy_score(y, test_y_predicted)
print(accuracy)
```
